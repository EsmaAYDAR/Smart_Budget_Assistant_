##  Projenin Amacı 
Kullanıcıdan alınan harcama tutarları ile çalışan,inputun istediğimiz değişken türünde olup olmadığını test eden ve ona göre output veren kısaca harcama alışkanlıklarımızla ilgili bilgiler verebilen **akıllı bütçe asistanı** yazılımıdır.

Günlük hayatta, muhasebe kalemleri veya günlük maliye yönetimi konususnda bizi yönlendirerek harcamalarımızı asiste eder.

Kullanılan Konular:

    Fonksiyonlar: Veri döndürme

    Args: Sınırsız veri giriş desteği

    Try-Except: Hata verdiğinde programanın çökmemesi

    Kapsam: Değişkenlerin ulaştığı yer

```python
def get_expenses():
    #kullanıcıya kaç tane harcama gireceğini sormamız gerek.
    liste=[] #boş liste oluşturuldu.
    adet=int(input("kaç harcama gireceksiniz?"))
    for i in range (adet): #belirtilen adet kadar döngü kurulur.
        try:
            a=float(input("HARCAMANIZI GİRİNİZ: "))
            liste.append(a)
        except:
            print("LÜTFEN HARCAMANIZI SAYIYLA GİRİN.")
    return liste        
```

```python
get_expenses()
```

```python
def calculate_summary(harcamalar):
    if not harcamalar:
        return 0, 0  #liste boş ise hata almasın.0 döndürsün.
    toplamHarcama=sum(harcamalar)
    ortalamaHarcama=round(toplamHarcama/len(harcamalar),2)

    return toplamHarcama,ortalamaHarcama    
```

```python
harcamalarim=[1.0,2.0,3.0]
toplamHarcama,ortalamaHarcama=calculate_summary(harcamalarim)
print("Toplam:", toplamHarcama)
print("Ortalama:", ortalamaHarcama)
```
